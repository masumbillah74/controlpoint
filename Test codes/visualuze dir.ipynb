```python
from matplotlib import pyplot as plt, patches
from matplotlib.collections import PatchCollection
import numpy as np
import random
from IPython.display import display, clear_output

length = 0.8
width = 7
height = 6
pm = np.array([+1, -1])

tag_coord = [1,1]

fig, ax = plt.subplots(1, 1, figsize=(7, 5.4))

tag_coord = [2,2]
rects = []

color = ["#"+''.join([random.choice('0123456789ABCDEF') for j in range(6)]) 
                      for i in range(360)]

k=0
for degree in range(0, 360, 45):
#for degree in [0, 45, 90]:
    x2, y2 = tag_coord[0], tag_coord[1]
    rect = patches.Rectangle((tag_coord[0], tag_coord[1]), length, length, edgecolor=None, alpha=.9, linewidth=0, angle=degree)
    coords = rect.get_patch_transform().transform(rect.get_path().vertices[:-1])
    x1, y1 = coords[3][0], coords[3][1]

    x = x1 + pm*(1.5*(x2-x1))
    y = y1 + pm*(1.5*(y2-y1))
    x = x[0] if np.round(abs(x[0]-x2), 4) <=length/2 else x[1]
    y = y[0] if np.round(abs(y[0]-y2), 4) <=length/2 else y[1]

    while True:
        rect = patches.Rectangle((x, y), length, length, #color=color[++k], 
                                 edgecolor=None, alpha=.1, linewidth=0, angle=degree)
        rects.append(rect)

        coords = rect.get_patch_transform().transform(rect.get_path().vertices[:-1])
        x = coords[1][0]
        y = coords[1][1]

        if min(coords[:,0])>width or max(coords[:,0])<0 or min(coords[:,1])>height or max(coords[:,1])<0:
            break
    k+=1
    
ax.add_collection(PatchCollection(rects, match_original=True))

tag_rect = patches.Rectangle((tag_coord[0], tag_coord[1]), 0.05, 0.05, color='red')
ax.add_patch(tag_rect)

ax.set_xlim([0, width])
ax.set_ylim([0, height])

plt.show()
```
